```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv('./wrangling/assignment/data/airbnb_hw.csv')
df_1.head(10)

```

```python

# Converted 'Price' to string type so that we can clean i
df_1['Price'] = df_1['Price'].astype(str)

# Removed any non-numeric characters and convert the type of the Pricecolumn to float
df_1['Price'] = df_1['Price'].str.replace('[^\d.]', '', regex=True)
```

```python
# coerced the column to being numeric
df_1['Price'] = pd.to_numeric(df_1['Price'], errors='coerce')

# Create a missing value dummy variable
df_1['price_nan'] = df_1['Price'].isnull().astype(int)

# Checked for any missing values after we cleaned the Price column
total_missing = df_1['price_nan'].sum()

print(total_missing)
```

```python
df_2 = pd.read_csv('./wrangling/assignment/data/sharks.csv')
df_2.head(10)

```

```python
# View the unique values in the 'Type' column

unique_types = df_2['Type'].unique()
print(unique_types)
```

```python
# Clean the 'Type' variable
df_2['Type'] = df_2['Type'].str.strip()  # Remove leading/trailing whitespace
df_2['Type'] = df_2['Type'].str.lower()  # Convert to lowercase for consistency
df_2['Type'] = df_2['Type'].str.replace(' ', '_')  # Replace spaces with underscores

# Replace missing values with NaN if needed
df_2['Type'] = df_2['Type'].replace(['', 'nan', 'NaN'], np.nan)
```

```python
# Further cleaning can be done based on specific requirements, such as grouping similar types together.
# For example:
df_2['Type'] = df_2['Type'].replace(['boat', 'boating', 'Boat', 'Boating', 'boatomg', 'watercraft'], 'boat_related')

# Check the unique values after cleaning
cleaned_unique_types = df_2['Type'].unique()
print(cleaned_unique_types)

```

```python
# changed all values for Type that arent unprovoked, provoked, or boat_related to 'unverified'

df_2['Type'] = np.where(df_2['Type'].isin(['unprovoked', 'provoked', 'boat_related']), df_2['Type'], 'unverified')
```

```python
cleaned_unique_types = df_2['Type'].unique()
print(cleaned_unique_types)
```
